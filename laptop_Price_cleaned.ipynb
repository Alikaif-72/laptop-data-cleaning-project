```python
df.drop(['Rating' , 'Reviews'] , axis=1 , inplace=True)
```

```python
df.fillna({'SSD': 'No SSD'}, inplace=True).head(3)
```

```python
df.fillna({'Core': df['Core'].mode()[0]}, inplace=True).head(5)
```

```python
df.fillna({'Core': df['Core'].mode()[0]}, inplace=True).head(5)
```

```python
df.duplicated().sum()
```

```python
df = df.dropna(subset=['Actual Price', 'Saving'])
```

```python
df.isnull().sum()
```

```python
df['extra_information'] = df['Name'].str.extract(r'\((.*?)\)')
df['Name'] = df['Name'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
df['Name'] = df.apply(
    lambda x: x['Name'].replace(x['Model'], '').strip(),
    axis=1
)
df['Name'] = df['Name'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
df.fillna({'extra_information':'Not Specified'} , inplace = True).head(5)
```

```python
col = df.pop('Brand')      # Column nikaal lo
df.insert(1, 'Brand', col) # Index 2 par insert karo
```

```python
df.drop('Name' , axis = 1 , inplace = True)
```

```python
df.rename(columns={'Unnamed: 0':'S.no',
                   'Brand':'Brand_Name'}, 
          inplace=True)
```

```python
df.rename(columns={
    'Discounted_Price': 'Discounted_Price (₹)',
    'Actual_Price': 'Actual_Price (₹)'
}, inplace=True)
```

```python
df.head(5)
```

```python
df.rename(columns={'S.no':'s_no',
                  'Brand_name':'brand_name',
                  'Discounted_Price (₹)':'discounted_price',
                  'Actual_Price (₹)':'actual_price',
                  'Saving':'saving',
                  'Core':'core',
                  'SSD':'ssd',
                  'Model':'model'
}, inplace=True)
```

```python
df['saving'] = df['saving'].str.replace('% OFF', '')
df['saving'] = df['saving'].astype(int)
```

```python
df.style.format({
    'discounted_price': '₹{:,.0f}',
    'actual_price': '₹{:,.0f}'
})
```
